```python
import pandas as pd

# Titanic veri setini yükle (Kaggle kernel)
df = pd.read_csv('/kaggle/input/titanic/train.csv')


```

## Kazada ölenlerin yaş ortalamasını bulunuz


```python
dead = df[df['Survived'] == 0]
dead_age_mean = dead['Age'].mean()
dead_age_mean
```

## Kazada ölenlerin bilet fiyatlarının ortalamasını ve medyanını bulunuz

```python
dead_fare_mean = dead['Fare'].mean()
dead_fare_median = dead['Fare'].median()
dead_fare_mean, dead_fare_median
```

## Kazada ölen erkeklerin yaş ortalamasını bulunuz

```python
dead_males = dead[dead['Sex'] == 'male']
dead_male_age_mean = dead_males['Age'].mean()
dead_male_age_mean
```

## Kazada ölen kadınların yaş ortalamasını bulunuz

```python
dead_females = dead[dead['Sex'] == 'female']
dead_female_age_mean = dead_males['Age'].mean()
dead_female_age_mean
```

##  Kazadan kurtulanların yaş ortalamasını bulunuz

```python
survived = df[df['Survived'] == 1]
survived_age_mean = survived['Age'].mean()
survived_age_mean
```

## Kazadan kurtulanların bilet fiyatlarının ortalamasını bulunuz

```python
survived_fare_mean = survived['Fare'].mean()
survived_fare_mean
```

 ## Kazadan kurtulan toplam kişi sayısını bulunuz

```python
total_survived = len(df[df['Survived'] == 1])
total_survived
```

 ## 10 yaşından küçüklerin bilet fiyatlarının medyan değerini bulunuz

```python
under_10 = df[df['Age'] < 10]
under_10_fare_median = under_10['Fare'].median()
under_10_fare_median
```

##  1, 2 ve 3. sınıf bilet fiyatlarının ortalama ve medyanlarını karşılaştırınız

```python
fare_stats = df.groupby('Pclass')['Fare'].agg(['mean', 'median'])
fare_stats
```

## Kazada ölen kadınların oranı ile erkeklerin oranını karşılaştırınız

```python
total_males = len(df[df['Sex'] == 'male'])
dead_males = len(df[(df['Sex'] == 'male') & (df['Survived'] == 0)])
male_death_ratio = dead_males / total_males

total_females = len(df[df['Sex'] == 'female'])
dead_females = len(df[(df['Sex'] == 'female') & (df['Survived'] == 0)])
female_death_ratio = dead_females / total_females

male_death_ratio, female_death_ratio
```
